# First half of the AO3 practical session

## This will focus on *developing research questions*

This session will answer:
- What kind of data are on AO3?

- What kind of rq's can we answer with this data? 

- What ethical considerations should we make when collecting fanfiction data?

### Exercise 1: Familiarizing yourself with AO3 (10 min)

Let's have a poke around.

If you are already familiar with AO3, consider:
- what are specific affordances and constraints of the platform (e.g., filtering, searching, tagging)?
 

### Exercise 2: Develop a research question (10 min, in groups)

After poking around on AO3, what things stick out to you? What ideas for RQ's do you get? 

### Exercise 3: Reflecting on methodology (10 min, in groups)

What data would you need to answer this RQ? 

How will you collect this data? What ethical considerations and measures do you need to make?
